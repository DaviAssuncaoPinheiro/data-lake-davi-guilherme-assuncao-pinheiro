# Análise de Modalidades\n\nRanking dos esportes olímpicos por quantidade de medalhas.

```python
import pandas as pd
import json
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path
from datetime import datetime
```

```python
BRONZE_DIR = Path("../../bronze")
OUTPUT_DIR = Path(".")

df_mod = pd.read_parquet(BRONZE_DIR / "modalidades_1986_2024.parquet")

summary = df_mod.groupby("sport")[["Gold", "Silver", "Bronze", "Total"]].sum()
summary = summary.sort_values("Total", ascending=False).reset_index()

summary.to_csv(OUTPUT_DIR / "modalidades_summary.csv", index=False)

print(f"Total de esportes: {len(summary)}")
summary.head(15)
```

```python
top15 = summary.head(15).set_index("sport")

fig, ax = plt.subplots(figsize=(12, 6))
top15[["Gold", "Silver", "Bronze"]].plot(
    kind="barh",
    stacked=True,
    color=["#FFD700", "#C0C0C0", "#CD7F32"],
    ax=ax
)
ax.set_xlabel("Quantidade de Medalhas")
ax.set_ylabel("")
ax.set_title("Top 15 Esportes por Medalhas")
ax.invert_yaxis()
plt.tight_layout()
plt.savefig(OUTPUT_DIR / "modalidades_plot.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
metadata = {
    "nome": "Análise de Modalidades",
    "fonte": "Camada bronze - modalidades_1986_2024",
    "descricao": "Ranking dos esportes olímpicos por medalhas",
    "formato": "csv/png",
    "colunas": [{"nome": col, "tipo": str(summary[col].dtype)} for col in summary.columns],
    "quantidade_linhas": len(summary),
    "data_coleta": datetime.now().strftime("%Y-%m-%d"),
    "observacoes": ""
}

with open(OUTPUT_DIR / "metadata.json", "w", encoding="utf-8") as f:
    json.dump(metadata, f, ensure_ascii=False, indent=4)
```
